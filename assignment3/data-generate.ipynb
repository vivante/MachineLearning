```python
from sklearn.datasets import make_blobs
from sklearn.preprocessing import normalize
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def make_data(num_samples):
    x, y = make_blobs(n_samples=num_samples, n_features=2, centers=2, cluster_std=3)
    dataset = np.concatenate((x, y.reshape((y.size, 1))), axis=1)
    return dataset
```

```python
def save_data(data, filename, fmt):
    '''Save as CSV'''
    np.savetxt(filename, data, delimiter=',', fmt=fmt)
```

```python
def plot_data(x, y):
    plt.scatter(x[:, 0], x[:, 1], c=y)
    plt.show()
```

```python
data = make_data(100)
plot_data(data[:, [0, 1]], data[:, -1])
```

```python
save_data(data, './data/dataset.temp.csv', ['%.18f', '%.18f', '%d'])
```
