```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('./clusters-3.csv')
xs = dataset.iloc[:].values
```

```python
x1, x2 = xs[:, 0], xs[:, 1]
plt.scatter(x1, x2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Data')
plt.show()
```

```python
class KMeans():
    def __init__(self, K):
        if (K <= 0):
            raise ValueError('K has to be greater than zero')
        
        self.K = K
        self.centroids = self.rand_centroids()  # centroids of the clusters (the index is the cluster number)
        self.assignments = None
        self.assignments_prev = None
        self.num_iter = 0
    
    def __str__(self):
        return 'Centroids:\n{}'.format(self.centroids)

    def rand_centroids(self):
        return np.array([ xs[i] for i in np.random.randint(0, len(xs), size=self.K) ])

    def fit(self):
        def distance_sq(a, b):
            return np.sum((a - b) ** 2)

        def mean_vec(xss):
            return (1 / len(xss)) * np.sum(xss, axis=0)

        def assign_cluster(x):
            return np.argmin([distance_sq(x, u) for u in self.centroids])

        def new_centroid(k):
            x_indices = np.where(self.assignments == k)[0]  # indices of data pts that are assigned to cluster 'k'
            xss = np.take(xs, x_indices, axis=0)
            if (len(xss) == 0):
                return xs[np.random.randint(0, len(xs))]
            return mean_vec(xss)
        
        # K-means Algorithm:
        self.centroids = self.rand_centroids()  # random init step
        self.assignments = np.array([ assign_cluster(x) for x in xs ])  # cluster assignment step
        while not np.array_equal(self.assignments, self.assignments_prev):
            self.assignments_prev = self.assignments.copy()
            self.centroids = np.array([ new_centroid(k) for k in np.arange(0, self.K) ])  # move centroid step
            self.assignments = np.array([ assign_cluster(x) for x in xs ])  # cluster assignment step
            self.num_iter += 1
        
        print('Converged in {} iterations'.format(self.num_iter))
```

```python
def plot_centroids(model):
    x1, x2 = xs[:, 0], xs[:, 1]
    plt.scatter(x1, x2, c=model.assignments)
    for u in model.centroids:
        plt.scatter(u[0], u[1], c='black', s=200, marker='X')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title('Clusters')
    plt.show()
```

```python
means3 = KMeans(3)
print(means3)
```

```python
plot_centroids(means3)
```

```python
means3.fit()
print(means3)
```

```python
plot_centroids(means3)
```
