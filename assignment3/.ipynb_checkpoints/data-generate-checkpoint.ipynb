```python
from sklearn.datasets import make_blobs
from sklearn.preprocessing import normalize
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def make_data():
    num_samples = 700
    x, y = make_blobs(n_samples=num_samples, n_features=2, centers=3, cluster_std=1.5)
    plot_data(x, y)
    dataset = np.concatenate((x, y.reshape((num_samples, 1))), axis=1)
    return dataset
```

```python
def save_data(data):
    ''' Save as CSV '''
    np.savetxt('dataset.temp.csv', data, delimiter=',', fmt=['%.18f', '%.18f'])
```

```python
def plot_data(x, y):
    plt.scatter(x[:, 0], x[:, 1], c=y)
    plt.show()
```

```python
data = make_data()
data
```

```python
clusters = np.delete(data, 2, 1)
```

```python
save_data(clusters)
```
