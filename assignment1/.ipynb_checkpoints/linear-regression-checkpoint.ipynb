```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.metrics import explained_variance_score
```

```python
# Import dataset
dataset = pd.read_csv('50Startups.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
dataset
```

```python
# Handle missing data
imputer = SimpleImputer(missing_values=0, strategy='mean')
x[:, :3] = imputer.fit_transform(x[:, :3])
```

```python
# Encode categorical data 
ct = ColumnTransformer([ ('one-hot', OneHotEncoder(categories='auto'), [-1]) ])
encoded_cols = ct.fit_transform(x)
x = np.concatenate((x[:, :-1], encoded_cols[:, :-1]), axis=1)  # non-categ cols ++ categ cols with dummy trap removed
encoded_cols[:, :-1].shape
```

```python
# Split data into training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=0)
```

```python
# Feature scaling
sc_x = StandardScaler()
sc_x.fit(x_train.astype('float64'))
x_train, x_test = sc_x.transform(x_train.astype('float64')), sc_x.transform(x_test.astype('float64'))
```

```python
# Augment the feature matrix
augment = np.ones((len(x_train), 1))
x_train = np.concatenate((augment, x_train), axis=1)
augment = np.ones((len(x_test), 1))
x_test = np.concatenate((augment, x_test), axis=1)
```

```python
class LinearRegression:
    def __init__(self, feature_len, learning_rate):
        self.params = np.random.randint(10, size=feature_len)
        self.learning_rate = learning_rate
        self.cost_tracker = None  # costs of the model over the previous time i ran gradient descent
    
    def __str__(self):
        return 'Theta: {}'.format(self.params)
    
    def h(self, x):
        return np.dot(self.params, x)
    
    def predict(self, xs):
        return np.array(list(map(lambda x: self.h(x), xs)))

    def fit(self, num_iter):
        costs = [] # keep track of the cost function through the iterations

        def step(ind_param):
            ''' Map from old param to new param '''
            param_index, param = ind_param[0], ind_param[1] 
            return param - self.learning_rate * self.cost_d(param_index)

        # The gradient descent algo
        for _ in range(num_iter):
            # Track the cost
            costs.append(self.cost())

            new_params = map(step, enumerate(self.params))
            self.params = np.array(list(new_params)) # update the params
        
        self.cost_tracker = costs

    def base_cost(self):
        def diff_sq(elem):
            x, y = elem[0], elem[1]
            return (self.h(x) - y) ** 2

        elems = map(diff_sq, zip(x_train, y_train))
        result = np.sum(list(elems))
        return result

    def base_cost_d(self, param_index):
        feature_index = param_index
        def diff_feature(elem):
            x, y = elem[0], elem[1]
            return (self.h(x) - y) * x[feature_index]

        elems = map(diff_feature, zip(x_train, y_train))
        result = 2 * np.sum(list(elems))
        return result
    
    def cost(self):
        ''' Cost function '''
        raise NotImplementedError('You must implement the cost function')

    def cost_d(self, param_index):
        ''' Partial derivative of the cost function w.r.t param at given index '''
        raise NotImplementedError('You must implement the cost function\'s partial derivative')
```

```python
class SimpleLinear(LinearRegression):
    def __init__(self, feature_len, learning_rate=0.5):
        super().__init__(feature_len, learning_rate)
    
    def cost(self):
        return self.base_cost()
    
    def cost_d(self, param_index):
        return self.base_cost_d(param_index)
```

```python
class LassoLinear(LinearRegression):
    def __init__(self, feature_len, learning_rate=0.5, regu=0.05):
        super().__init__(feature_len, learning_rate)
        self.regu = regu
    
    def cost(self):
        penalty = self.regu * np.sum(np.absolute(self.params))
        return self.base_cost() + penalty
    
    def cost_d(self, param_index):
        penalty_d = self.regu * np.sign(self.params[param_index])
        return self.base_cost_d(param_index) + penalty_d
```

```python
class RidgeLinear(LinearRegression):
    def __init__(self, feature_len, learning_rate=0.5, regu=0.05):
        super().__init__(feature_len, learning_rate)
        self.regu = regu
    
    def cost(self):
        penalty = self.regu * np.sum(self.params ** 2)
        return self.base_cost() + penalty
    
    def cost_d(self, param_index):
        penalty_d = self.regu * 2 * self.params[param_index]
        return self.base_cost_d(param_index) + penalty_d
```

```python
feature_length = len(x_train[0])
alpha1, alpha2 = 0.005, 0.01
regu = 0.05
reg1 = SimpleLinear(feature_length, alpha1)
reg2 = LassoLinear(feature_length, alpha1, regu)
reg3 = RidgeLinear(feature_length, alpha1, regu)
reg4 = SimpleLinear(feature_length, alpha2)
reg5 = LassoLinear(feature_length, alpha2, regu)
reg6 = RidgeLinear(feature_length, alpha2, regu)
```

```python
reg1.fit(100)
reg2.fit(100)
reg3.fit(100)
reg4.fit(100)
reg5.fit(100)
reg6.fit(100)
```

```python
# plt.xlim(0, 10)
plt.ylim(1.4e10, 1.6e10)
plt.plot(reg1.cost_tracker, color='blue', linestyle='dashed', label='Simple')
plt.plot(reg2.cost_tracker, color='red', linestyle='dashed', label='Lasso')
plt.plot(reg3.cost_tracker, color='green', linestyle='dashed', label='Ridge')
plt.title("Cost function with learning rate 0.005")
plt.xlabel("No. of iterations")
plt.ylabel("Cost function")
plt.legend(loc='upper right')
plt.show()
```

```python
# plt.xlim(0, 10)
plt.ylim(1.4e10, 1.6e10)
plt.plot(reg4.cost_tracker, color='blue', linestyle='dashed', label='Simple')
plt.plot(reg5.cost_tracker, color='red', linestyle='dashed', label='Lasso')
plt.plot(reg6.cost_tracker, color='green', linestyle='dashed', label='Ridge')
plt.title("Cost function with learning rate 0.01")
plt.xlabel("No. of iterations")
plt.ylabel("Cost function")
plt.legend(loc='upper right')
plt.show()
```

```python
# Measure the accuracy
preds1 = reg1.predict(x_test)
preds2 = reg2.predict(x_test)
preds3 = reg3.predict(x_test)
preds4 = reg4.predict(x_test)
preds5 = reg5.predict(x_test)
preds6 = reg6.predict(x_test)
acc1 = explained_variance_score(y_test, preds1) * 100
acc2 = explained_variance_score(y_test, preds2) * 100
acc3 = explained_variance_score(y_test, preds3) * 100
acc4 = explained_variance_score(y_test, preds4) * 100
acc5 = explained_variance_score(y_test, preds5) * 100
acc6 = explained_variance_score(y_test, preds6) * 100
```
