```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
```

```python
# Import dataset
dataset = pd.read_csv('Social_Network_ads.csv')
x = dataset.iloc[:, [2, 3]].values  # get only age and estimated salary features
y = dataset.iloc[:, -1].values
dataset
```

```python
# Split data into training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3, random_state=0)
```

```python
# Feature scaling
sc = StandardScaler()
sc.fit(x_train.astype('float64'))
x_train, x_test = sc.transform(x_train.astype('float64')), sc.transform(x_test.astype('float64'))
```

```python
# Fit the SVM
svm = SVC(kernel='linear', random_state=0)
svm.fit(x_train, y_train)
```

```python
# Test the model
predictions = svm.predict(x_test)
cm = confusion_matrix(y_test, predictions)
```

```python
X_set, y_set = x_train, y_train
X_grid_0 = np.arange(start=X_set[:, 0].min() - 1, stop=X_set[:, 0].max() + 1, step=0.01)
X_grid_1 = np.arange(start=X_set[:, 1].min() - 1, stop=X_set[:, 1].max() + 1, step=0.01)
X1 , X2 = np.meshgrid(X_grid_0, X_grid_1)

X3 = svm.predict(np.array([X1.ravel(),X2.ravel()]).T).reshape(X1.shape)

plt.contourf(X1, X2, X3, alpha=0.50, cmap=ListedColormap(("yellow", "cyan")))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(
        X_set[y_set == j, 0],
        X_set[y_set == j, 1],
        label = j,
        edgecolors = "Black"
    )
plt.title("Social Network Ads (Linear kernel)")
plt.xlabel("Age")
plt.ylabel("EstimatedSalary")
plt.legend()
plt.show()
```
