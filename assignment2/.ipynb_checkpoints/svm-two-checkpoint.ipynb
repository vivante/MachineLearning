```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
```

```python
# Import dataset
dataset1 = pd.read_csv('Social_Network_ads.csv')
x1 = dataset1.iloc[:, [2, 3]].values  # get only age and estimated salary features
y1 = dataset1.iloc[:, -1].values

dataset2 = pd.read_csv('Admission_Predict.csv')
x2 = dataset2.iloc[:, :-1].values
y2 = dataset2.iloc[:, -1].values
```

```python
# Split data into training and testing
x_train1, x_test1, y_train1, y_test1 = train_test_split(x1, y1, test_size=1/3, random_state=0)
x_train2, x_test2, y_train2, y_test2 = train_test_split(x2, y2, test_size=0.25, random_state=0)
```

```python
# Feature scaling
sc1 = StandardScaler()
sc1.fit(x_train1.astype('float64'))
x_train1, x_test1 = sc1.transform(x_train1.astype('float64')), sc1.transform(x_test1.astype('float64'))

sc2 = StandardScaler()
sc2.fit(x_train2.astype('float64'))
x_train2, x_test2 = sc2.transform(x_train2.astype('float64')), sc2.transform(x_test2.astype('float64'))
```

```python
# Fit the SVMs
svm1 = SVC(kernel='rbf', random_state=0)
svm1.fit(x_train1, y_train1)

svm2 = SVC(kernel='rbf', random_state=0)
svm2.fit(x_train2, y_train2)
```
